### WHAT DOES A ML ALGORITHM CONSIST OF

Machine learning is nothing but mathematical optimization. We use a <b>model</b> to <b>fit</b> data, i.e associate features to target/labels. Consider an example where we want to determine the price of a house based on it's area. We have some data that gives the area of a house and the corresponding price. We just try to fit a <b>model</b> which in this case is a line through this data, such that the <b>cost</b> is minimal. Then we input our <b>features</b> i.e the area of the house in the <b>hypothesis</b> that is the optimal equation of the line to get the <b>target</b> which in this case is the price.
<center></center>

###  Given a dateset of change of water level in reservoir and amount of water flowing out, predict the amount of water ﬂowing out of a dam using the change of water level in a reservoir.

This kind of a machine learning problem involves regression, whrerin we have to predict values

```python
import numpy as np   ## as we have aalready seen numpy helps with numerical calculations
import matplotlib.pyplot as plt  ## matplotlib helps to plot
import pandas as pd  ## pandas helps us to import, export and query data.
## loading the important libraries
```

```python
data=pd.read_csv('Dataset.csv')
data

## reading data from the .csv file into a data frame, data types are automatically assigned by pandas.
## To see how .csv (comma seperated values) file stores data you can open it using any text editor
```

```python
X=data.iloc[:,:1]
## splitting data int features and labels. X is the features and y the label i.e. what we want to predict.
y=data.iloc[:,1:]
X.head()
## head() returns the first 5 rows
```

```python
## Visualizing the data-set
## plotting the dataset
plt.figure(figsize=(8,5))  ## determines the size of the figure
plt.ylabel('Water flowing out of the dam (y)') ## determines the label at the y-axis
plt.xlabel('Change in water level') ## determines the label at the x-axis
plt.plot(X,y,'rx') ## plotting data
plt.grid() ## shows a grid
plt.show()
```

It is an important feature of machine learning to divide data into 2 parts, training data and test data. Training data 
is used to train a model and test data is used to test it. And the test data is used to see how well the data is trained. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
## The data is divided randomly into training set and test set,by default is a 75,25 split
plt.figure(figsize=(8,5))
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level')
plt.grid()
plt.plot(X_train,y_train,'rx') ## plotting the training data
## Now try plotting the test data
plt.plot(X_test,y_test,'bx')
plt.show()
```

#### Try modifying test_size and random_state to see what happens

```python
from sklearn import linear_model ## model to fit data
from sklearn.metrics import mean_squared_error   ## Import fn to calculate cost
```

```python
reg = linear_model.LinearRegression() ## determining the model
```

```python
reg.fit(X_train,y_train) ## fitting the training data, to get the hypothesis
```

```python
plt.plot(X_train,y_train,'rx') ## plotting X_train and y_train
plt.scatter(X_test,reg.predict(X_test),color = 'blue') ## plotting y_pred for X_test the result of hypothesis fn.
plt.scatter(X_test,y_test,color = 'green')
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level')
plt.grid()
plt.show()
## blue line shows how data is fit, data is fit linearly
## green line shows the actual value for X_test
```

### ARE WE DONE YET ?

We can see from the graph above, that the blue dots are not a very accurate representation of the output (green dots). This is because we have tried to develop a linear relationship between X and y, however the actual relationship appears to be of a higher order. Such situations are called underfitting. When we are underfitting the data even less training data gives a similar output, this is because the features used to estimate are not good enough. This is where using polynomial features comes in.

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(12)
## generating polynomial powers of x and using them as new data points
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 2,test_size = 0.3) ## splitting into testing and training
X_train_poly = poly.fit_transform(X_train) ## adds polynomial features of all degrees from 0 to selected value
X_test_poly = poly.fit_transform(X_test) ## adds polynomial features of all degrees from 0 to selected value
print(X_train)
print(X_train_poly)
## You can see the relation between X_train and X_train_poly
```

```python
reg.fit(X_train_poly,y_train) ## fitting new data
```

```python
plt.figure(figsize=(13,10))
plt.plot(X_train,y_train,'rx') ## actual data points
plt.scatter(X_train,reg.predict(X_train_poly), color = 'blue') 
plt.scatter(X_test,reg.predict(X_test_poly),color = 'green')
plt.ylabel('Water flowing out of the dam (y)')
plt.xlabel('Change in water level')
plt.grid()
plt.show()
## try using different degrees of polynomials and see the nature of the curve
```

Trying very high powers like 12 for example with a decent test_set size of 30%, we realize that while the higher powers try to fit the training data too perfectly, looking at the local trends, they sure do miss out on the global trends. This is called overfitting. When a model overfits training acuuracy becomes high but the test accuracy becomes low. While training data we need to ensure that it neither underfits nor overfits but needs to have the perfect fit. In this case we can plot a graph of y vs X, but we cannot do this when we have more number of features. So this is where learning curves come in

### THE LEARNING CURVE

<ul><li>Graph that compares the performance of a model on training and testing data over a varying number of training instances</li>
<li>We should generally see performance improve as the number of training points increases</li>
<li>When we separate training and testing sets and graph them individually<ul>
    <li>We can get an idea of how well the model can generalize to new data</li></ul></li>
<li>Learning curve allows us to verify when a model has learning as much as it can about the data</li>
<li>When it occurs<ul>
<li>The performances on the training and testing sets reach a plateau</li>
    <li>There is a consistent gap between the two error rates</li></ul></li>
<li>The key is to find the sweet spot that minimizes bias and variance by finding the right level of model complexity</li>
    <li>Of course with more data any model can improve, and different models may be optimal</li>
</ul>   
<center>
</center>

When we have high variance :<ul>
    <li>The gap between training and testing error is high.</li>
    <li>The gap appers to be reducing if we increase the no. of training examples.</li>
    <li>This occurs, if the hypothesis is too complex</li><br>
    

```python
poly = PolynomialFeatures(4)
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 5,test_size = 0.3)
X_train_poly = poly.fit_transform(X_train) ## adds polynomial features of all degrees from 0 to selected value
X_test_poly = poly.fit_transform(X_test)
```

```python
## Drawing the learning curve to see how well the algorithm has learnt
X_train_error = [] ## list that will be storing training errors for training sets of diff. sizes
X_test_error = [] ## list that will be storing test errors for training sets of diff. sizes
train_set_size = np.arange(1,len(X_train) + 1,1) ## making a counting list from 1,2,3,... , size of X_train
from sklearn.metrics import mean_squared_error
for i in train_set_size: ## itertaing
    X_lc_tr = X_train_poly[0:i] ## making small training sets of sizes ranging from 1 to X_train
    y_lc_tr = y_train[0:i] ## making small training sets of sizes ranging from 1 to y_train
    reg.fit(X_lc_tr,y_lc_tr) ## fitting data on those small sets
    pred_test = reg.predict(X_test_poly) ## predicting output on test set
    pred_train = reg.predict(X_lc_tr)  ## predicting output on training set
    X_train_error.append(mean_squared_error(y_lc_tr,pred_train)) ## appending errors to array
    X_test_error.append(mean_squared_error(y_test,pred_test))
    plt.scatter(X,y,color = 'red') ## all data points
    plt.scatter(X_test,pred_test,color = 'blue') 
    plt.scatter(X_lc_tr[:,1],pred_train,color = 'green') 
    plt.show()
    ## Uncoment the aabove lines to see how learning takes place, they will show how learning 
    #takes place as the training set size increases
plt.figure(figsize=(13,7))
plt.plot(train_set_size,X_train_error)
plt.plot(train_set_size,X_test_error) 
plt.ylim((0,50))
plt.xlabel('Training set size')
plt.ylabel('Error')
plt.title('Learning curve')
plt.grid()
plt.show()
```

#### Looking at learning curves we can predict whether our model is underfitting, overfitting or has perfect fit. <br>
If the model is underfitting<ul>
    <li>We need to find better features that well represent the target.</li>
    <li>Use more polynomial features</li>
    <li>Reduce the regularization parameter  (We will see later)</li></ul>
If the model is overfitting<ul>
    <li>Reduce the number of features i.e non-contributing features</li>
    <li>Reduce degree of polynomial features</li>
    <li>Increse no. of data points</li>
    <li>Increase the regularization parameter</li>


To find the best degree of polynomial features, we plot the degrees vs errors

```python
## testing for different degrees of polynomials
from sklearn import preprocessing as pre
degrees = [] ## making an empty list that stores
min = 1  ## Enter the min value
max = 15    ## Enter max value here
error_test = []   
error_train = []
degrees = np.arange(min,max+1,1)
for x in degrees:  # Iterating over degrees
    poly = PolynomialFeatures(x)
    X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 9,test_size = 0.2)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.fit_transform(X_test)
    scaler = pre.StandardScaler().fit(X_train)  # Scaling features to mean 0 and variance 1.
    X_train_poly =  scaler.transform(X_train_poly)
    X_test_poly = scaler.transform(X_test_poly)
    reg.fit(X_train_poly,y_train)  ## fitting data
    error_test.append(mean_squared_error(y_test,reg.predict(X_test_poly)))
    error_train.append(mean_squared_error(y_train,reg.predict(X_train_poly)))
plt.figure(figsize=(15,10))
plt.plot(degrees,error_test,'r')
plt.plot(degrees,error_train,'g') 
plt.ylim((0,50))
plt.xlabel('Max. degree')
plt.ylabel('Error')
plt.show()    
```

From the above graph we can see that in the beginning both the test  and train errors are high and the model is underfiiting, then both the errors become low and almost converge, thus we can say at that point that the model is quiet good. When the degree of polynomial increases, test error shoots up and at this point the model is overfitting

### IMPROVING PERFORMANCE FURTHER

Though we have selected the perfect no. of polynomial features, there is still a pretty significant gap between the training errors and test errors. This gap must be reduced. Reduction of this gap means that the samples that the model hasn't trained on will have similar performance to the training sample. We can attain this by using regularization 

Consider the hypothesis $\theta_3x^3+\theta_2x^2+\theta_1x+\theta_0$ to be overfitting. If we reduce the coefficient $a$ significantly, the nature of the hypothesis becomes more quadratic. Hence we can come up with a new and better cost function that makes sure the coefficients are not allowed to rise to get more accuracy on training data.<br>
$cost = \sum_{i=1}^m (h_\theta(x^i) - y^i)^2 + \lambda \sum_{i=1}^n \theta_i^2$<br>
$\lambda$ is the regularization parameter and controls the extent of regularization

#### RIDGE REGRESSION IS THE REGRESSION MODEL WITH THE MODIFIED COST FUNCTION

```python
## So 4 is giving us best results, now moving on to ridge regression
from sklearn.linear_model import Ridge
```

```python
alphas = np.linspace(0,10,300)
error_test = []
error_train = []
for item in alphas: ## iterating over diff values of regularization parameter
    X_train, X_test, y_train, y_test = train_test_split(X, y,random_state = 5,test_size = 0.2) ## splitting
    poly = PolynomialFeatures(4) ## setting polynomial features based on above results 
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.fit_transform(X_test)
    scaler = pre.StandardScaler().fit(X_train) ## scaling features
    X_train_poly =  scaler.transform(X_train_poly)
    X_test_poly = scaler.transform(X_test_poly)
    reg = Ridge(alpha = item) ## trying diff regularization parameters
    reg.fit(X_train_poly,y_train)
    error_test.append(mean_squared_error(y_test,reg.predict(X_test_poly)))
    error_train.append(mean_squared_error(y_train,reg.predict(X_train_poly)))
plt.plot(alphas,error_test,'r')
plt.plot(alphas,error_train,'g') 
plt.ylim((0,50))
plt.show()
```

### LEARNING CURVE WITH REGULARIZATION

```python
## Drawing the learning curve to see how well the algorithm has learnt
X_train_error = [] ## list that will be storing training errors for training sets of diff. sizes
X_test_error = [] ## list that will be storing test errors for training sets of diff. sizes
train_set_size = np.arange(1,len(X_train) + 1,1) ## making a counting list from 1,2,3,... , size of X_train
from sklearn.metrics import mean_squared_error
for i in train_set_size: ## itertaing
    X_lc_tr = X_train_poly[0:i] ## making small training sets of sizes ranging from 1 to X_train
    y_lc_tr = y_train[0:i] ## making small training sets of sizes ranging from 1 to y_train
    reg = Ridge(1)
    reg.fit(X_lc_tr,y_lc_tr) ## fitting data on those small sets
    pred_test = reg.predict(X_test_poly) ## predicting output on test set
    pred_train = reg.predict(X_lc_tr)  ## predicting output on training set
    X_train_error.append(mean_squared_error(y_lc_tr,pred_train)) ## appending errors to array
    X_test_error.append(mean_squared_error(y_test,pred_test))
    plt.scatter(X,y,color = 'red') ## all data points
    plt.scatter(X_test,pred_test,color = 'blue') 
    plt.scatter(X_train,reg.predict(X_train_poly),color = 'green') 
    plt.show()
    ## Uncoment the aabove lines to see how learning takes place, they will show how learning 
    #takes place as the training set size increases
plt.figure(figsize=(13,7))
plt.plot(train_set_size,X_train_error)
plt.plot(train_set_size,X_test_error) 
plt.ylim((0,50))
plt.xlabel('Training set size')
plt.ylabel('Error')
plt.title('Learning curve')
plt.grid()
plt.show()
```
